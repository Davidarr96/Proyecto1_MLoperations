```python
#Vamos a importar las librerias a utilizar 
import pandas as pd
import json
from pandas.io.json import json_normalize
import ast
import numpy as np
```

```python
# Ingresamos los datasets y guardamos cada uno con una variable distinta
movies_df = pd.read_csv('/content/drive/MyDrive/Datasets_MLoperations/movies_dataset.csv',sep = ',', encoding = 'utf-8')
```

# **TRANSFORMACIONES**






```python
#Utilizamos una funcion para poder desanidar los datos dentro de los dicc y las listas
def fetch_name(obj): 
    if isinstance(obj, str) and '{' in obj:
        L=[]
        for i in ast.literal_eval(obj):
            L.append(i['name']);
        return L
    
def fetch_name2(obj): 
    
    if isinstance(obj, str) and '{' in obj:
        # print(obj)
        dic = ast.literal_eval(obj)

        return dic['name']
    

    

movies_df['genres'] = movies_df['genres'].apply(fetch_name)
movies_df['belongs_to_collection'] = movies_df['belongs_to_collection'].apply(fetch_name2)
movies_df['production_companies'] = movies_df['production_companies'].apply(fetch_name)
movies_df['production_countries'] = movies_df['production_countries'].apply(fetch_name)
movies_df['spoken_languages'] = movies_df['spoken_languages'].apply(fetch_name)
movies_df
```

```python
# Rellenar los valores nulos de 'revenue' y 'budget' con el número 0
movies_df['revenue'] = movies_df['revenue'].fillna(0)
movies_df['budget'] = movies_df['budget'].fillna(0)
```

```python
# Los valores nulos del campo relase_date deben eliminarse
movies_df = movies_df.dropna(subset=['release_date'])
```

```python
# Convertir la columna release_date a formato fecha
movies_df['release_date'] = pd.to_datetime(movies_df['release_date'], errors='coerce')

# Crear la columna release_year con el año de estreno
movies_df['release_year'] = movies_df['release_date'].dt.year

# Formatear la columna release_date como AAAA-mm-dd
movies_df['release_date'] = movies_df['release_date'].dt.strftime('%Y-%m-%d')
```

```python
#Creamos la columna llamada "return" dividiendo la columna "revenue" por la columna budget.
#Para poder realizarlo correctamente, como primera condicion nos aseguramos que los datos sean del mismo tipo.
movies_df['revenue'] = movies_df['revenue'].astype(float)

```

```python
#Encontramos valores no numeros en el campo "budget", para solucionar esto convertimos esos str a 0.
# Definir una función que convierta los valores a float
def convertir_a_float(valor):
    if isinstance(valor, str):
        # Si el valor es un string, intentar convertirlo a float
        try:
            return float(valor)
        except ValueError:
            # Si no se puede convertir, devolver cero
            return 0.0
    else:
        # Si el valor no es un string, devolverlo sin cambios
        return valor

# Aplicar la función al campo 'budget'
movies_df['budget'] = movies_df['budget'].apply(convertir_a_float)
```

```python
#Una vez que los campos tienen el mismo tipo, creamos la columna "return"

# reemplazar valores nulos por 0
movies_df['revenue'] = movies_df['revenue'].fillna(0)
movies_df['budget'] = movies_df['budget'].fillna(0)

# definir la función para calcular el retorno de inversión
def calculate_return(row):
    if row['budget'] == 0:
        return 0
    else:
        return row['revenue'] / row['budget']

# crear la columna 'return' utilizando la función apply
movies_df['return'] = movies_df.apply(calculate_return, axis=1)

```

```python
#Eliminamos las columnas que no vamos a utilizar:[video, imdb_id, adult, original_title, vote_count, poster_path, homepage]
movies_df = movies_df.drop(['video', 'imdb_id', 'adult', 'original_title', 'vote_count', 'poster_path', 'homepage'], axis=1)
```
